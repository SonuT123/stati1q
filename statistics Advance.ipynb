Q3. Write a Python function to calculate the probability density function of a normal distribution with
     given mean and standard deviation at a given point.

```python
import math
def normal_pdf(x, mean, std):
    exponent = -((x - mean) ** 2) / (2 * std ** 2)
    coefficient = 1 / (std * math.sqrt(2 * math.pi))
    return coefficient * math.e ** exponent


print(normal_pdf(0, 0, 1)) 
print(normal_pdf(1, 0, 1)) 
print(normal_pdf(2, 0, 1)) 


"""f(x)=σ2π​1​e−2σ2(x−μ)2​
where x is the variable, μ is the mean, and σ is the standard deviation.
The function imports the math module to use the constant e and the square root function.
It then defines the function with three parameters: x, mean, and std. The function calculates the exponent of
e by using the formula, and then multiplies it by the coefficient of e, which is also calculated by using the formula.
The function returns the product of coefficient and e raised to the exponent, which is the probability density value at x."""
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
    and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size = 1000  
p = 0.4             

random_sample = np.random.binomial(n=1, p=p, size=sample_size)

plt.hist(random_sample, bins=2, edgecolor='black') 
plt.title(f'Binomial Distribution (n=1, p={p}) - Random Sample of Size {sample_size}')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['Failure', 'Success']) 
plt.show()



"""
>We set sample_size to 1000, which is the size of the random sample we want to generate.
The probability of success p is set to 0.4.

>We use np.random.binomial to generate the random sample. The n parameter is set to 1 because we are 
simulating a binomial distribution with only one trial per data point.

>We create a histogram of the random sample using Matplotlib, specifying the number of bins as 2 
(for the two possible outcomes: 0 and 1)."""
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
     with given mean at a given point.

```python
import scipy.stats as stats

def poisson_cdf(k, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution at a given point.

    Args:
    k (int): The point at which to calculate the CDF.
    mean (float): The mean (average) of the Poisson distribution.

    Returns:
    float: The CDF value at the given point.
    """
    cdf_value = stats.poisson.cdf(k, mu=mean)
    return cdf_value

mean = 3.5
point = 2
cdf_at_point = poisson_cdf(point, mean)
print(f"CDF at k = {point}: {cdf_at_point}")


"""
>mean is the mean (average) of the Poisson distribution.
>k is the point at which you want to calculate the CDF.
>The stats.poisson.cdf function is used to calculate the CDF at the specified point k, with the specified mean mu.
"""
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
    sample mean and variance.

```python
import numpy as np

mean = 5
sample_size = 1000

random_sample = np.random.poisson(mean, sample_size)

sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
"""
We import the numpy library as np.
>We set the mean of the Poisson distribution (mean) and the desired sample size (sample_size).
>We use np.random.poisson(mean, sample_size) to generate a random sample of size 1000 from a Poisson distribution with the specified mean.
>We calculate the sample mean using np.mean(random_sample) and the sample variance using np.var(random_sample)."""
```
